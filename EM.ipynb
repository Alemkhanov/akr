```python
import numpy as np
from math import factorial
```

```python
y = np.array([7, 7, 1, 3, 2, 1, 5, 1, 4, 0, 2, 3, 3, 1, 2, 2, 3, 1, 4, 0, 8, 3, 5, 3, 0, 0, 7, 1, 1, 3, 1, 3, 2, 4, 1, 6, 2, 2, 4, 1, 3, 1, 1, 2, 7, 3, 3, 2, 2, 2]) 
M = 10
eps = 1e-6
iteration = 0
x = np.random.rand()
Pi_0 = np.array([x,1-x])
Mu_0 = np.random.rand(2)
Pi_1 = np.array([100,100])
Mu_1 = np.array([100,100])
```

```python
def factor(m,i,j):
    a = factorial(m) // (factorial(i) * factorial(m-i))
    return a * j ** i * (1-j) ** (m-i)
```

```python
while np.abs(Pi_1-Pi_0).sum() + np.abs(Mu_1-Mu_0).sum() > eps:
    print('iteration = ' + str(iteration), ':\n' , 'Pi = ' + str(Pi_0), '\n' , 'Mu = ' + str(Mu_0), end='\n\n')
    Pi_1 = Pi_0
    Mu_1 = Mu_0
    r = []
    for i in y:
        b = []
        for j in Mu_0:
            b.append(factor(M,i,j))
        r.append(b)
    r = np.array(r)
    r *= Pi_0[None,:]
    r /= np.sum(r, axis = 1)[:,None]
    Pi_0 = np.average(r, axis = 0)
    Mu_0 = np.sum(y[:, None] * r, axis = 0) / (r.sum(axis = 0) * M)
    iteration += 1
```
